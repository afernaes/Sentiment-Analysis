## Summary

The objective of this project relies on doing a sentiment analysis for a client to compare the sentiment between 5 types of smart phones focusing in iPhone and Galaxy. To do that, we mapped around 800 websites from Common Crawl using AWS and then created a matrix with 58 features. These features contains the number of times the ‘positive’, ‘neutral’ or ‘negative’ words were found with some characteristics from these smart phones.
Here we analyze the overall sentiment towards Iphone and Samsung Galaxy, aiming to recommned one the devices.

We received two data sets where the sentiment was labeled manually. These datasets contains results from 12973 websites, while the sentiment is categorized from 0 (very negative) to 5 (very positive).

## Dependant Variable Modification

- We decided to modify the dependant variable, since from our perspective, negative, neutral and positive are enought classifiers for sentiment, thus helping to reduce the uncertainity

```python
# Recode the sentiment
print("Iphone Sentiment counts before recoding:", iphone_data.iphonesentiment.value_counts())

iphone_data['iphonesentiment'] = iphone_data['iphonesentiment'].replace([0, 1, 2, 3, 4, 5],
                                            ['Negative', 'Negative', 'Neutral', 'Neutral', 'Positive', 'Positive'])

print("Galaxy Sentiment counts before recoding:", galaxy_data.galaxysentiment.value_counts())

galaxy_data['galaxysentiment'] = galaxy_data['galaxysentiment'].replace([0, 1, 2, 3, 4, 5],
                                            ['Negative', 'Negative', 'Neutral', 'Neutral', 'Positive', 'Positive'])
```

```python
# Convert to factor
iphone_data.iphonesentiment = iphone_data.iphonesentiment.astype('category')
#Counts after recoding
print("Iphone Sentiment counts before recoding\n", iphone_data.iphonesentiment.value_counts())

galaxy_data.galaxysentiment = galaxy_data.galaxysentiment.astype('category')
#Counts after recoding
print("Galaxy Sentiment counts before recoding\n", galaxy_data.galaxysentiment.value_counts())
```

```python
#I create here a function, otherwise it removes the columnames

def variance_threshold_selector(data, threshold):
    selector = VarianceThreshold(threshold)
    selector.fit(data)
    return data[data.columns[selector.get_support(indices=True)]]

iphone_ROS_nzv=variance_threshold_selector(iphone_ROS.iloc[:,0:58],threshold=0.1)
iphone_ROS_nzv['iphonesentiment']=iphone_ROS['iphonesentiment']

galaxy_ROS_nzv=variance_threshold_selector(galaxy_ROS.iloc[:,0:58],threshold=0.1)
galaxy_ROS_nzv['galaxysentiment']=galaxy_ROS['galaxysentiment']
```

### Iphone

#### UpSampling

```python
# Separate train set and test set
X=iphone_ROS.iloc[:,0:58]
y=iphone_ROS['iphonesentiment']
X_train, X_validation, Y_train, Y_validation = train_test_split(X, y,
                                                                test_size=0.20,
                                                                random_state=1,
                                                                shuffle=True)
```

```python
models = []

models.append(('KNN', KNeighborsClassifier()))
models.append(('SVM', SVC()))
models.append(('RF', RandomForestClassifier()))
#models.append(('XGB', GradientBoostingClassifier()))


# Evaluate each model in turn
names = []
results = []

for name, model in models :
    cv_results = cross_val_score(model, X_train, Y_train, cv=5, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    model.fit(X_train, Y_train)
    predictions = model.predict(X_validation)
    print('%s: CV accuracy: %f std: (%f)' % (name, cv_results.mean(), cv_results.std()))
    print('Test accuracy: %f' % (accuracy_score(Y_validation, predictions)))
    print('Test Kappa: %f' % (cohen_kappa_score(Y_validation, predictions)))
    
# Compare Algorithms
plt.boxplot(results, labels=names)
plt.title('Algorithm Comparison')
plt.show()
```

#### Modelling Tuning

```python
# Separate train set and test set
X=iphone_ROS_corr.iloc[:,0:47]
y=iphone_ROS_corr['iphonesentiment']
X_train, X_validation, Y_train, Y_validation = train_test_split(X, y,
                                                                test_size=0.20,
                                                                random_state=1,
                                                                shuffle=True)
```

```python
#DEFINE THE PARAMETER GRID FOR GRID SEARCH CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1500, num = 15)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 100, num = 7)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Initiate classifier
rf = RandomForestClassifier()
#Randomized Grid Search CV
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 30, cv = 4, verbose=0, random_state=42, n_jobs = 1)
#The most important arguments in RandomizedSearchCV are n_iter, which controls the number of different combinations to try, and cv which is the number of folds to use for cross validation (we use 100 and 3 respectively).
# Fit model with the best RF found
rf_random.fit(X_train, Y_train)
# Predict values for validation data
prediction = rf_random.predict(X_validation) 
```

```python
print('Accuracy:', (accuracy_score(Y_validation, prediction)))
print('Kappa:', (cohen_kappa_score(Y_validation, prediction)))
print('Best parameters:', (rf_random.best_params_))
```

#### Modelling Tuning

```python
# Separate train set and test set
X=galaxy_ROS_corr.iloc[:,0:48]
y=galaxy_ROS_corr['galaxysentiment']
X_train, X_validation, Y_train, Y_validation = train_test_split(X, y,
                                                                test_size=0.20,
                                                                random_state=1,
                                                                shuffle=True)
```

```python
#DEFINE THE PARAMETER GRID FOR GRID SEARCH CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1500, num = 15)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 100, num = 7)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Initiate classifier
rf = RandomForestClassifier()
#Randomized Grid Search CV
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 30, cv = 4, verbose=0, random_state=42, n_jobs = 1)
#The most important arguments in RandomizedSearchCV are n_iter, which controls the number of different combinations to try, and cv which is the number of folds to use for cross validation (we use 100 and 3 respectively).
# Fit model with the best RF found
rf_random.fit(X_train, Y_train)
# Predict values for validation data
prediction = rf_random.predict(X_validation) 
```

```python
print('Accuracy:', (accuracy_score(Y_validation, prediction)))
print('Kappa:', (cohen_kappa_score(Y_validation, prediction)))
print('Best parameters:', (rf_random.best_params_))
```
